```python
#201901881_설희관
import csv

f = open('gender.csv')
data = csv.reader(f)

size = []

province = input("도 이름을 입력하세요")
county = input("구 이름을 입력하세요")

for row in data:
    if province and county in row[0]:
        m = 0
        f = 0
        for i in row[3:104]:
            m = m + (int(i.replace(',','')))
        for i in row[106:]:
            f = f + (int(i.replace(',','')))
        break
        
size.append(m)
size.append(f)

import matplotlib.pyplot as plt

label = ['male', 'female']
color = ['skyblue', 'hotpink']
explodeEffect = [0.1, 0] #explode 처리

plt.axis('equal')
plt.rc('font', family = 'Malgun Gothic')
plt.title('April_2020_Population_' + province + county + ' 지역 남녀 성별 인구 분포') #title 표시
plt.pie(size, labels = label, autopct = '%.2f%%', colors = color, radius = 1, startangle = 90, explode = explodeEffect, shadow = True) #pie chart, 소수점 아래 2자리까지 표시
plt.legend(title = province + ' ' + county, loc = 'center right', bbox_to_anchor = (1, 0, 0.5, 1)) #legend 표시
plt.show()
```
