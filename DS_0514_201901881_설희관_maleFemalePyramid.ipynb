```python
#201901881_설희관
import csv

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
county = input('구 이름을 입력하세요:  ')
name = input('읍면동 이름을 입력하세요:  ')

for row in data:
    if (county and name) in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',',''))) #slicing
        for i in row[106:]:
            f.append(int(i.replace(',',''))) #slicing
            
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False #음수값 오류 처리
plt.title('April_2020_Population_' + county + ' ' + name + '_MyHometown_pyramidChart') #차트 제목
plt.xlabel('cnt') #x축 제목
plt.ylabel('age') #y축 제목
plt.barh(range(101), m, color = 'blue', label = 'male')
plt.barh(range(101), f, color = 'pink', label = 'female')
plt.legend()
plt.show()
```
