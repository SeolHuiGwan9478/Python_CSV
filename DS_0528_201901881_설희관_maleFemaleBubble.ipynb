```python
#201901881_설희관
import csv
import math

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
size = []

province = input("도 이름을 입력하세요 ")
county = input("구 이름을 입력하세요 ")

for row in data:
    if (province and county) in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i + 103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',','')) + int(row[i + 103].replace(',',''))))
        break #원의크키는 인구수의 제곱근으로 표현

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
#title
plt.title("April_2020_Population_" + province + county + ' 연령대별 남성_여성 Bubble_Chart')
plt.xlabel('남성인구수') #xlabel
plt.ylabel('여성인구수') #ylabel
#colormap 'rainbow' 사용
plt.scatter(m, f, c = range(101), cmap = 'rainbow', alpha = 0.5) #투명도 alpah로 처리
plt.plot(range(max(m)), range(max(m)), 'g')
plt.colorbar()
plt.show()
```
